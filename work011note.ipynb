```python
import numpy as np
import matplotlib.pylab as plt
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
```

```python
def ref_torus(P1, P2, D, d, mode):
    if mode == 'A':
        X = (D + d*np.cos(P1)) * np.cos(P2) + D/2
        Y = d * np.sin(P1)
        Z = (D + d*np.cos(P1)) * np.sin(P2)
    if mode == 'B':
        X = (D+d*np.cos(P1)) * np.cos(P2) - D/2
        Y = (D+d*np.cos(P1)) * np.sin(P2)
        Z = d*np.sin(P1)
    return (X,Y,Z)
```

```python
N = 2**5
P1 = np.linspace(0,1,N) * 2*np.pi
P2 = np.linspace(0,1,N) * 2*np.pi
PP1, PP2 = np.meshgrid(P1, P2)
X,Y,Z = ref_torus(PP1, PP2, 1.0, 0.3, mode='A')
X2,Y2,Z2 = ref_torus(PP1, PP2, 1.0, 0.3, mode='B')
```

```python
fig = plt.figure(figsize=[20/2.57, 20/2.57])
fig.clf()
ax = Axes3D(fig)
# ax.plot_wireframe(X, Y, Z, cmap='Reds', alpha=.5, color='gray')
# ax.plot_wireframe(X2, Y2, Z2, cmap='Blues', alpha=.5, color='b')
ax.plot_surface(X, Y, Z, cmap='Reds', alpha=.5, color='gray')
ax.plot_surface(X2, Y2, Z2, cmap='Blues', alpha=.5, color='b')
ax.set_xlim([-1.5, 1.5])
ax.set_ylim([-1.5, 1.5])
ax.set_zlim([-1.5, 1.5])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
#ax.view_init(30, 60)
#plt.close(fig)
```

```python
def gen_torus(P1, P2, D, d, mode):
    if mode == 'A':
        X = (D + d*np.cos(P1)) * np.cos(P2) 
        Y = d * np.sin(P1) + D
        Z = (D + d*np.cos(P1)) * np.sin(P2)
    if mode == 'B':
        X = (D + d*np.cos(P1)) * np.cos(P2) 
        Y = d * np.sin(P1) - D
        Z = (D + d*np.cos(P1)) * np.sin(P2)
    return (X,Y,Z)
```

```python
N = 2**5
P1 = np.linspace(0,1,N) * 2*np.pi
P2 = np.linspace(0,1,N) * 2*np.pi
PP1, PP2 = np.meshgrid(P1, P2)
X,Y,Z = gen_torus(PP1, PP2, 1.0, 0.3, mode='A')
X2,Y2,Z2 = gen_torus(PP1, PP2, 1.0, 0.3, mode='B')
```

```python
fig = plt.figure(figsize=[20/2.57, 20/2.57])
fig.clf()
ax = Axes3D(fig)
# ax.plot_wireframe(X, Y, Z, cmap='Reds', alpha=.5, color='gray')
# ax.plot_wireframe(X2, Y2, Z2, cmap='Blues', alpha=.5, color='b')
ax.plot_surface(X, Y, Z, cmap='Reds', alpha=.5, color='gray')
ax.plot_surface(X2, Y2, Z2, cmap='Blues', alpha=.5, color='b')
ax.set_xlim([-1.5, 1.5])
ax.set_ylim([-1.5, 1.5])
ax.set_zlim([-1.5, 1.5])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
#ax.view_init(30, 60)
#plt.close(fig)
```
